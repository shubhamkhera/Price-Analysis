### 2.2 Choosing 3 pairs of colums to find relation by visualising

#### Wheel Type Vs Price

```python
dfall['drive-wheels'].value_counts()
```

```python
plt.figure(figsize=(10,7))
sns.set_theme(style="darkgrid")
plt1 = sns.boxplot(x="drive-wheels", y="price", hue = "drive-wheels", data=dfall,width = .6, dodge=False)
plt1.set(xlabel= "Wheel Type", ylabel = "Price", title = "Wheel Type Vs Price")
```

#### Engine-size Vs Price

```python
plt.figure(figsize=(14,7))
sns.set_theme(style="darkgrid")
plt2 = sns.regplot(x="engine-size", y="price", data=dfall) #shows the relation between Engine size and prices
plt2.set(xlabel= "Engine Size", ylabel = "Price", title = "Scatterplot of Engine-size Vs Price")
plt.ylim(0,)
plt.xlim(0,)
```

```python
plt.figure(figsize=(14,7))
sns.set_theme(style="darkgrid")
plt2 = sns.scatterplot(x="city-mpg", y="price", data=dfall, hue="drive-wheels") #shows the relation between City Milage and Engine size
plt2.set(xlabel= "Milage", ylabel = "Price", title = "City Milage Vs Price")
```
